```
#upload file for this session
from google.colab import files
uploaded = files.upload()
```

```
#path of the file to read
flight_filepath = "/content/flight_delays.csv"

#read the file into a variable flight_df
flight_df = pd.read_csv(flight_filepath, index_col="Month")
```

```
#examine the data
print(flight_df.shape)
flight_df
```

### Bar Chart

```
#bar chart
plt.figure(figsize=(10, 6))

plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

sns.barplot(x=flight_df.index, y=flight_df['NK'])

plt.ylabel("Arrival delay (in minutes)")
```

* sns.barplot - This tells the notebook that we want to create a bar chart.

* x=flight_df.index - This determines what to use on the horizontal axis. In this case, we have selected the column that indexes the rows (in this case, the column containing the months).

* y=flight_df['NK'] - This sets the column in the data that will be used to determine the height of each bar. In this case, we select the 'NK' column.


**Important Note**<br>
x=flight_df['Month'] - This will return an error. "Month" column was used to index the *rows* when you loaded the dataset.<br>
When selecting the indexing column, you have to make a notation as **flight_df.index**

### Heatmap

```
plt.figure(figsize=(14, 7))

plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap(data=flight_df, annot=True)

plt.xlabel("Airline")
```

```
#upload data for this notebook
print("Here, we're working with insurance.csv data.")
from google.colab import files
uploaded = files.upload()
```

```
ins_filepath = "/content/insurance.csv"

ins_df = pd.read_csv(ins_filepath)
ins_df.head()
```

```
#let's scatter plot
sns.scatterplot(x=ins_df['bmi'], y=ins_df['charges'])
```

From the plot result above, it suggests that bmi and insurance charges are quite positively correlated.<br>
Let's check if it's true by simply plotting a regression line.

```
sns.regplot(x=ins_df['bmi'], y=ins_df['charges'])
```

Great! Now you can see that they are actually positively correlated.<br>

Now let's see if we can explore the relationships between more than two variables.

```
sns.scatterplot(x=ins_df['bmi'], y=ins_df['charges'], hue=ins_df['smoker'])
```

The graph above is the result of color coding the points by 'smoker'.<br>
Looks like those who smoke are tend to pay more on insurance than those who don't.<br>

**sns.lmplot** gives you regression line plots to check the correlation for each pattern.

```
sns.lmplot(x="bmi", y="charges", hue="smoker", data=ins_df)
```

There is one more interesting plot you can find on seaborn's scatter plot.<br>
**sns.swarmplot** will give you interesting plots on categorical variables.

```
sns.swarmplot(x=ins_df['smoker'], y=ins_df['charges'])
```

This **categorical scatter plot** above tells us that
* on average, non-smokers are charged less than smokers and
* those who pay the most are smokers; whereas those who pay the least are non-smokers.

* **winpercent** column contains values which show how popular that candy is.

```
#Which is popular, 3 Musketeers or Almond Joy?
winpercent_3M = candy_df[candy_df["competitorname"]=="3 Musketeers"]["winpercent"]
winpercent_AJ = candy_df[candy_df["competitorname"]=="Almond Joy"]["winpercent"]
print("3 Musketeers : ", winpercent_3M.values[0])
print("Almond Joy : ", winpercent_AJ.values[0])
```

By scatter plotting the dataset, you can see if people tend to prefer candies with higher sugar content.

Could you tell any correlation between the two features?<br>
To confirm your thought, let's plot a regression line in the scatter plot.

```
plt.figure(figsize=(8, 6))

sns.regplot(x=candy_df["sugarpercent"], y=candy_df["winpercent"])
plt.title("Relationship between sugar percent and the popularity(winpercent)")
```

By plotting a regression line, you can slightly see a positive correlation between the two features.
So you might be a bit more confident to say that people tend to prefer candies with more sugars.<br>




Let's take a further look another relationship between the "pricepercent" and the "winpercent" also by color coding the "chocolate" feature.

```
plt.figure(figsize=(8, 6))

sns.scatterplot(x=candy_df["pricepercent"], y=candy_df["winpercent"], hue=candy_df["chocolate"])
plt.title("Relationship between sugar percent and the popularity(winpercent)")
```

Now with regression line.

```
plt.figure(figsize=(8, 6))

sns.lmplot(x="pricepercent", y="winpercent", hue="chocolate", data=candy_df)
plt.title("Relationship between sugar percent and the popularity(winpercent)")
```

You can see from this plot that even though the price of the candies with chocolate become higher, they tend to become more popular accordingly. In the case of non-chocolate candies, they have negative correlation.

## Histograms and Density plots

```
#upload file for this session (iris.csv)
from google.colab import files
uploaded = files.upload()
```

**iris.csv**<br>
Each row in the dataset corresponds to a different flower. There are four measurements: the sepal length and width, along with the petal length and width. We also keep track of the corresponding species.

```
iris_filepath = "/content/iris.csv"

iris_df = pd.read_csv(iris_filepath, index_col="Id")
iris_df.head()
```

### Histogram
**sns.distplot**

```
sns.distplot(a=iris_df['Petal Length (cm)'], kde=False)
```

* **a=** chooses the column we'd like to plot (in this case, we chose 'Petal Length (cm)').
* **kde=False** is something we'll always provide when creating a histogram, as leaving it out will create a slightly different plot.

### KDE Plot
KDE(kernel density estimate) plot is like a smoothed histogram.
**sns.kdeplot** command will provide you with its image.

```
sns.kdeplot(data = iris_df['Petal Length (cm)'], shade=True )
```

### 2D-KDE plot
2-dimension KDE plot will give you an image of the kde plot over two different features.
**sns.jointplot**
<br>
The darker the area, more likey we are to observe those combination of two features in the dataset.

```
sns.jointplot(x=iris_df['Petal Length (cm)'], y=iris_df['Sepal Width (cm)'], kind="kde")
```
